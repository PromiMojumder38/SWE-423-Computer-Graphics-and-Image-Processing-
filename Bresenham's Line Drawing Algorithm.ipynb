```python
import matplotlib.pyplot as plt

def drawline(x0, y0, x1, y1):
    points = []
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)

    # Case 0 < m < 1
    if dx >= dy:
        slope_error_new = dy * 2 - dx
        slope_error_old = slope_error_new
        y = y0
        for x in range(x0, x1 + 1):
            points.append((x, y))
            slope_error_old = slope_error_new
            if slope_error_old >= 0:
                y += 1
                slope_error_new -= dx * 2
            slope_error_new += dy * 2
    # Case m > 1
    else:
        slope_error_new = dx * 2 - dy
        slope_error_old = slope_error_new
        x = x0
        for y in range(y0, y1 + 1):
            points.append((x, y))
            slope_error_old = slope_error_new
            if slope_error_old >= 0:
                x += 1
                slope_error_new -= dy * 2
            slope_error_new += dx * 2

    return points

# Case 1: (1,1), (8,4)
x0, y0 = 1, 1
x1, y1 = 8, 4
points1 = drawline(x0, y0, x1, y1)

# Case 2: (1,1), (4,8)
x0, y0 = 1, 1
x1, y1 = 4, 8
points2 = drawline(x0, y0, x1, y1)

# Plotting lines among the points
plt.figure(figsize=(10, 5))

# Plotting Case 1
plt.subplot(1, 2, 1)
plt.title('Case 1: (1,1), (8,4)')
plt.plot([point[0] for point in points1], [point[1] for point in points1], color='blue')
plt.plot([points1[0][0], points1[-1][0]], [points1[0][1], points1[-1][1]], color='red')
plt.scatter([point[0] for point in points1], [point[1] for point in points1], color='#e6870b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

# Plotting Case 2
plt.subplot(1, 2, 2)
plt.title('Case 2: (1,1), (4,8)')
plt.plot([point[0] for point in points2], [point[1] for point in points2], color='#143314')
plt.plot([points2[0][0], points2[-1][0]], [points2[0][1], points2[-1][1]], color='#F50CA0')
plt.scatter([point[0] for point in points2], [point[1] for point in points2], color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)

plt.tight_layout()
plt.show()
```
